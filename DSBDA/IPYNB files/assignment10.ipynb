```python
import pandas as pd
import numpy as np
```

```python
dataset=pd.read_csv("D:\SEM - 6\DSBDA\Datasets\TitanicDataset.csv")
dataset.head()
```

```python
dataset.info()
```

```python
dataset.isnull().sum()
```

```python
avg_age=dataset['Age'].mean(axis=0)
print(avg_age)
dataset['Age'].replace(np.nan,avg_age,inplace=True)
```

```python
dataset['Cabin'].value_counts().idxmax()
dataset['Cabin'].replace(np.nan,"B57 B59 B63 B66",inplace=True)
```

```python
dataset['Fare'].replace(np.nan,dataset['Fare'].mean(),inplace=True)
```

```python
dataset.isnull().sum()
```
